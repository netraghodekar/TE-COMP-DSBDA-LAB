```python
import nltk
para =  'For me, becoming isn’t about arriving somewhere or achieving a certain aim. I see it instead as forward motion, a means of evolving, a way to reach continuously toward a better self. The journey doesn’t end.'
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
sent = sent_tokenize(para)
```

```python
sent[1]
```

```python
sent[0]
      
```

```python
words = word_tokenize(para)
words
```

```python
from nltk.corpus import stopwords
```

```python
swords = stopwords.words('english')
swords
```

```python
x = [word for word in words if word.lower() not in swords]
x
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem('working')
```

```python
y = [ps.stem(word) for word in x]
y
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wnl = WordNetLemmatizer()
```

```python
import nltk
nltk.download('omw-1.4')
```

```python
wnl.lemmatize('working', pos='v')

# a - adjective
# n- noun
# r - adverb
```
